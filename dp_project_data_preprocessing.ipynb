```python
import os
import pickle

# Define the root directory of your chest X-ray dataset
dataset_root = "/content/drive/MyDrive/chest_xray_dataset"

# Define the labels and corresponding folder names
labels = {
    "NORMAL": 0,
    "PNEUMONIA": 1
}

# Initialize the datapair list
datapair = []

# Iterate over the labels and subfolders
for label_name, label_value in labels.items():
    subfolder_path = os.path.join(dataset_root, label_name)
    if os.path.isdir(subfolder_path):
        for filename in os.listdir(subfolder_path):
            if filename.endswith(".jpg") or filename.endswith(".jpeg"):
                datapair.append((label_name, filename, label_value))

# Save datapair to a file with the 'root' key included
datapair_file_path = "/content/drive/MyDrive/chest_xray_dataset/datapair.dat"
with open(datapair_file_path, "wb") as fp:
    pickle.dump({"data_pair": datapair, "root": dataset_root}, fp)

print("Datapair file saved successfully as datapair.dat.")
```
